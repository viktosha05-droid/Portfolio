```python
import pandas as pd

# Upload files

df = pd.read_csv('survey_results_public.csv', encoding='utf-8', low_memory=False)

df_questions = pd.read_csv('survey_results_schema.csv', encoding='utf-8', low_memory=False)

# Check result
print(f"upload lines: {len(df)}")
df.head()
```

```python
pd.set_option('display.max_columns', None)
```

Task 1

```python
# Count unique values ResponseID
total_respondents_unique = df['ResponseId'].nunique()

# Output results
total_respondents_unique
```

Task 2

```python
# 1. List QName for schema
schema_questions = set(df_questions['qname'])

# 2. List all df columns
data_columns = set(df.columns)

# 3. Intersection: only identical columns
relevant_questions = list(schema_questions.intersection(data_columns))

# 4. Create subset
df_subset = df[relevant_questions]

# 5. Check for complete responses
# dropna() removes rows with any NaN
respondents_with_all_answers = df_subset.dropna()

# 6. Final_count
final_count = len(respondents_with_all_answers)

# Print final_count
print(f"Respondents_with_all_answers: {final_count}")
print(f"We checked the answers to {len(relevant_questions)} questions.")
```

Результат 0 респондентів є логічним для даного датасету, оскільки опитування містить 126 специфічних запитань, які залежать від попередніх відповідей, тому жоден респондент не заповнив абсолютно всі поля.

Task 3

```python
# Calc mean, median, mode
mean_exp = df['WorkExp'].mean()
median_exp = df['WorkExp'].median()
mode_exp = df['WorkExp'].mode()[0] 

# Create summary table
stats_work_exp = pd.DataFrame({
    'Metric': ['Mean', 'Median', 'Mode'],
    'Value': [mean_exp, median_exp, mode_exp]
})
# Display the table
stats_work_exp
```

Task 4

```python
# Sum all True values 'Remote'
remote_count = (df['RemoteWork'] == 'Remote').sum()
# Display result
print(f"Remote workers: {remote_count}")
```

Task 5

```python
# Check for 'Python' (ignore NaNs) and create mask
python_mask = df['LanguageHaveWorkedWith'].str.contains('python', case=False, na=False)
python_users = python_mask.sum()
# Count only non-empty responses
total_ans = df['ResponseId'].nunique()

# Calculate and format percentage
pct_python = round((python_users / total_ans) * 100, 2)
print (total_ans)
print (python_users)
print(f"Python popularity: {pct_python}%")
```

Task 6

```python
# Count online learners
# Column 'LearnCode' contains multiple values separated by semicolons
online_learners = df['LearnCode'].str.contains('Online Courses', na=False).sum()

# Display the final number
print(f"Number of online learners: {online_learners}")
```

Task 7

```python
# Reuse the Python mask from Task 5
python_df = df[python_mask]

# Group by filtered data
country_stats = python_df.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).dropna()
country_stats['mean'] = country_stats['mean'].round(2)
country_stats['median'] = country_stats['median'].round(0)
# Rename for better look
country_stats.columns = ['Mean Salary', 'Median Salary']
# Display result
country_stats.sort_values(by='Median Salary', ascending=False).head(10)
```

Task 8

```python
# ascending=False means from highest to lowest
top_5 = df.sort_values(by='ConvertedCompYearly', ascending=False).head(5)
# Display top 5 education levels
top_5[['ConvertedCompYearly', 'EdLevel']]
```

Task 9

```python
# Create a temporary column in the main df to mark Python users
# Use the mask from Task 5
df['PythonUser'] = python_mask

# Group by Age and calculate the percentage of Python users
python_by_age = df.groupby('Age')['PythonUser'].mean() * 100

# Convert to DataFrame and reset index for a clean look
python_by_age_df = python_by_age.reset_index()
python_by_age_df.columns = ['Age Group', 'Python Popularity (%)']

# Display results
python_by_age_df
```

Task 10

```python
# 1. Calculate 75th percentile for compensation
q75 = df['ConvertedCompYearly'].quantile(0.75)
# 2. Filter: high earners + remote workers
high_paid_remote = df[
    (df['ConvertedCompYearly'] >= q75) & 
    (df['RemoteWork'] == 'Remote')
]
# 3. Count top industries
# Most popular industries for top-tier remote roles
top_industries = high_paid_remote['Industry'].value_counts().head(10)
# Display result
top_industries
```
